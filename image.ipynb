```python
from PIL import Image

# Open the image
image_path = "diagram.png"
img = Image.open(image_path)
width, height = img.size

# Determine the split point
# Visually estimating the header (Title + Subtitle) takes up the top portion
# before "Phase 1" starts.
# Roughly 14% of the image height seems to cover the title and subtitle without cutting the Phase headers.
split_point = int(height * 0.14)

# Crop the header (top part)
header = img.crop((0, 0, width, split_point))

# Crop the body (rest of the image)
body = img.crop((0, split_point, width, height))

# Create a new image with the same dimensions
new_img = Image.new('RGB', (width, height), (242, 246, 252)) # Using a light background color matching the image

# Paste the body at the top (0, 0)
new_img.paste(body, (0, 0))

# Paste the header at the bottom
new_img.paste(header, (0, height - split_point))

# Save the result
output_path = "rearranged_architecture.png"
new_img.save(output_path)
```
